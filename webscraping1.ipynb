```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
```

```python
chrome_options= Options()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument("--disable-setuid-sandbox")
chrome_options.add_argument("--user-data-dir=/home/saumya/.config/google-chrome/Default")
browser = webdriver.Chrome(executable_path='/home/saumya/Downloads/chromedriver_linux64/chromedriver')
browser.get("https://news.google.com/topics/CAAqJggKIiBDQkFTRWdvSUwyMHZNRGx6TVdZU0FtVnVHZ0pKVGlnQVAB?hl=en-IN&gl=IN&ceid=IN%3Aen")
titles_element = browser.find_elements_by_xpath("//h3[@class='ipQwMb ekueJc gEATFF RD0gLb']")
titles = [x.text for x in titles_element]
print('titles:')
print(titles, '\n')
# sub_titles_element = browser.find_elements_by_xpath("//a[@class='DY5T1d']")
# sub_titles = [x.text for x in sub_titles_element]
# print('sub titles:')
# print(sub_titles, '\n')
sub_titles_element = browser.find_elements_by_xpath("//h4[@class='ipQwMb ekueJc gEATFF RD0gLb']")
subtitles = [x.text for x in sub_titles_element]
sub_titles=[]
for x in subtitles:
    if(x!=''):
        sub_titles.append(x)
print('sub titles:')
print(sub_titles, '\n')
print(len(titles))
print(len(sub_titles))
```

```python
#preprocessing our titles and subtitles
import string
titles = [x.strip() for x in titles]
sub_titles = [x.strip() for x in sub_titles]
titles = [x.lower() for x in titles]
sub_titles = [x.lower() for x in sub_titles]
def remove_punctuation(string): 
    punctuations = '''!|()-[]-{}_;:'"\,<>./?@#$%^&*_~'''
    for x in string: 
        if x in punctuations: 
            string = string.replace(x, "") 
    return string
titles = [remove_punctuation(x) for x in titles]
sub_titles = [remove_punctuation(x) for x in sub_titles]
print('titles:')
print(titles, '\n')
print('sub titles:')
print(sub_titles, '\n')
```
